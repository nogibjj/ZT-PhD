```python
import pandas as pd

# Read Excel file into DataFrame
df = pd.read_excel('Rushka.xlsx')

# Display DataFrame
print(df.head())
```

```python
df
```

```python
y = df["Реальные доходы\nнаселения\nКыргызстана\n(USD)"]
y
```

```python
x = df[["Разница ВВП между РУ и КР (доллар США)", "численность\nмигрантов РУ из КР (человек)", "Денежные переводы ( долларов США)"]]
x
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
mlr = LinearRegression()
mlr.fit(x, y)
```

```python
# intercept and coefficients

print("Intercept: ", mlr.intercept_)
print("Coefficients:")
list(zip(x, mlr.coef_))
```

```python
from sklearn import metrics

print('R squared: {:.2f}'.format(mlr.score(x, y)*100))
```
